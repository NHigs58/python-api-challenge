```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress
from config import api_key
from citipy import citipy
```

```python
# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

```python
# set up lists to hold reponse info
new_city = []
cloudiness = []
country = []
date = []
humidity = []
temp = []
lat = []
lng = []
wind = []
```

```python
record_counter = 0
set_counter = 1
# Starting URL for Weather Map API Call
# Save config information.
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "metric"

# Build partial query URL
query_url = f"{url}appid={api_key}&units={units}&q="


# Build partial query URL
print('------------------------')
print('Beginning Data Retrieval')
print('------------------------')

for city in cities:
    # Get weather data
    
    if record_counter < 50:
        record_counter += 1
    else:
        set_counter += 1
        record_counter = 0

    print('Processing record {} of set {} | {}'.format(record_counter, set_counter, city))
    try:
        response = requests.get(query_url + city).json() 
        cloudiness.append(response['clouds']['all'])
        country.append(response['sys']['country'])
        date.append(response['dt'])
        humidity.append(response['main']['humidity'])
        temp.append(response['main']['temp_max'])
        lat.append(response['coord']['lat'])
        lng.append(response['coord']['lon'])
        wind.append(response['wind']['speed'])
        new_city.append(city)
    except Exception as e: print(e)
        
    pass

print('-------------------------')
print('Data Retrieval Complete')
```

```python
# create a data frame

weather_dict = {
  "City": new_city,
    "Cloudiness" : cloudiness,
    "Country" : country,
    "Date" : date,
    "Humidity" : humidity,
    "Temp": temp,
    "Lat" : lat,
    "Lng" : lng,   
    "Wind Speed" : wind}

weather_pd = pd.DataFrame(weather_dict)
weather_pd.head()
```

```python
# Exporting to CSV
weather_pd.to_csv("weather_pd.csv")
```

```python
# Create Northern and Southern Hemisphere DataFrames
northern_hemisphere = weather_pd.loc[weather_pd["Lat"]>0.01]
southern_hemisphere = weather_pd.loc[weather_pd["Lat"]<-0.01]

# northern_hemisphere.head()
southern_hemisphere.head()
```

```python
# Northern Hemisphere - Max Temp vs. Latitude Linear Regression
x_values = northern_hemisphere['Lat']
y_values = northern_hemisphere['Temp']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"red")
plt.annotate(line_eq,(40,40),fontsize=15,color="red")

plt.ylim(-50,100)
plt.xlim(0, 80)

plt.ylabel("Max. Temp")
plt.xlabel("Latitude")
# plt.show()

plt.savefig("NorthMaxTempVsLat.png")
```
